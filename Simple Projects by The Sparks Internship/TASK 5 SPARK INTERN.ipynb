```python
merged_dataset['player_of_match'].value_counts().head(10).plot(kind = 'bar',figsize=(8,4), color='c')
plt.title("Top 10 players with Good Performance")
plt.ylabel("Frequency")
plt.xlabel("Players")
```

```python
#let's plot the top 10 run getter so far in IPL
```

```python
merged_dataset.groupby('batsman')['batsman_runs'].sum().sort_values(ascending = False).head(10).plot(kind = 'bar', color = 'c',
                                                                                            figsize = (15,5))
plt.title("Top Run Getters of Matches")
plt.xlabel("Batsmen")
plt.ylabel("Total Runs Scored")
plt.show()
```

```python
consistent_batsman = merged_dataset[merged_dataset.batsman.isin(['SK Raina', 'V Kohli','RG Sharma','G Gambhir',
                                            'RV Uthappa', 'S Dhawan','CH Gayle', 'MS Dhoni',
                                            'DA Warner', 'AB de Villiers'])][['batsman','season','total_runs']]

consistent_batsman.groupby(['season','batsman'])['total_runs'].sum().unstack().plot(kind = 'box', figsize = (10,6))
plt.title("Most Consistent batsmen of Matches")
plt.xlabel("Batsman")
plt.xticks(rotation=90)
plt.ylabel("Total Runs Scored each season")
plt.show()
```

```python
#We will consider players who have played 10 or more seasons
no_of_balls = pd.DataFrame(merged_dataset.groupby('batsman')['ball'].count()) #total number of matches played by each batsman
runs = pd.DataFrame(merged_dataset.groupby('batsman')['batsman_runs'].sum()) #total runs of each batsman
seasons = pd.DataFrame(merged_dataset.groupby('batsman')['season'].nunique()) #season = 1 implies played only 1 season

batsman_strike_rate = pd.DataFrame({'balls':no_of_balls['ball'],'run':runs['batsman_runs'],'season':seasons['season']})
batsman_strike_rate.reset_index(inplace = True)

batsman_strike_rate['strike_rate'] = batsman_strike_rate['run']/batsman_strike_rate['balls']*100
highest_strike_rate = batsman_strike_rate[batsman_strike_rate.season.isin([10,11])][['season','batsman','strike_rate']].sort_values(by = 'strike_rate',
                                                                                                           ascending = False)

highest_strike_rate.head(10)
```
